```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_excel('Online retail.xlsx',header=None)
df.head()
```

```python
pd.set_option('display.max_rows',None)
df[df.duplicated(keep=False)].head(100)
```

```python
# removing the duplicates
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
# keeping the items purchased in each transaction in a list
df=df.apply(lambda x:x.str.split(','))
df.head()
```

```python
# transformint the data which suits to apply TransactionEncoder
store=[i for i in df[0]]
```

```python
te=TransactionEncoder()
arry=te.fit_transform(store)
```

```python
arry[1]
```

```python
df=pd.DataFrame(arry,columns=te.columns_)
```

```python
df.head()
```

```python
# selecting the frequent items in the transactions with minimum support = 0.009 
frequent_itemsets = apriori(df, min_support=0.009, use_colnames=True)
frequent_itemsets
```

```python
# filtering the association rules having lift >1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
# mineral water & spaghetti are consequent for every antecedent items
rules[rules.confidence>=0.2]
```

# Association Rule Mining - Key Concepts

## 1. What is Lift and Why is it Important?

**Lift** is a measure that helps evaluate the strength of an association rule by comparing how much more often the items in the rule occur together than if they were statistically independent.

### Lift Formula:

\[
\text{Lift}(X \Rightarrow Y) = \frac{\text{P}(X \cap Y)}{\text{P}(X) \times \text{P}(Y)}
\]

- If **Lift > 1**: This implies a **positive correlation** between X and Y, meaning they occur together more often than expected.
- If **Lift < 1**: This implies a **negative correlation**, meaning they occur together less frequently than expected.
- If **Lift = 1**: X and Y are independent.

### Importance:
Lift helps in determining whether the association rule is simply a coincidence (due to the frequency of individual items) or if there is a true relationship between them. Higher lift values suggest stronger associations.

---

## 2. What is Support and Confidence?

### a. **Support**
Support indicates how frequently the itemsets appear in the dataset. It is defined as the proportion of transactions in which the itemset occurs.

#### Support Formula:

\[
\text{Support}(X) = \frac{\text{Transactions containing } X}{\text{Total Transactions}}
\]

### b. **Confidence**
Confidence measures the likelihood of occurrence of the consequent (Y) given that the antecedent (X) has occurred. It helps assess the reliability of the rule.

#### Confidence Formula:

\[
\text{Confidence}(X \Rightarrow Y) = \frac{\text{Support}(X \cap Y)}{\text{Support}(X)}
\]

### Example:
Consider the rule: **{Milk} → {Bread}**

- Support({Milk, Bread}): How frequently Milk and Bread are bought together.
- Confidence({Milk} → {Bread}): Given that a customer buys Milk, how often do they also buy Bread?

---

## 3. Challenges and Limitations of Association Rule Mining

1. **Combinatorial Explosion**:
   - The number of possible itemsets grows exponentially with the number of items, leading to computational inefficiency.
   
2. **Rare Item Problem**:
   - Rare items with small support values may not appear in frequent itemsets even if they have strong associations with other items.

3. **High Confidence Doesn’t Mean Causality**:
   - A high confidence score doesn't necessarily imply causality; it just shows correlation. Additional analysis is needed to determine the real relationship.

4. **Difficulties with Continuous Data**:
   - Association rule mining typically works well with categorical data, but handling continuous data requires discretization, which may lead to a loss of information.

5. **Threshold Sensitivity**:
   - Setting the thresholds for support and confidence can be tricky. Too high a threshold may result in missing interesting rules, while too low a threshold might generate too many uninteresting or noisy rules.
   
---

## Conclusion

- **Lift**, **Support**, and **Confidence** are crucial metrics for evaluating association rules.
- Challenges such as scalability, rare item problems, and threshold sensitivity make association rule mining a task that requires careful tuning and interpretation.
